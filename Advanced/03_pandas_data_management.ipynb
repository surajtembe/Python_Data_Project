```python
# copy() method
# when we create a copy of a df using copy method, the both df has different unique id's, 
# however when we assign a df to a variable (ex: df1 = df2), the df1 & df2 will has the same Id and if we make changes in any df other will also get altered

dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
df_altered = df_original.copy() #.copy() method
```

```python
print('Id of Original df: ', id(df_original))
print('Id of Altered df: ', id(df_altered))
print('Are the two df same?: ', id(df_original) ==id(df_altered))
```
